```python
from google.cloud import secretmanager

# Secret Manager setup remains the same
secret_id = "external-dataset-eff"
secret_client = secretmanager.SecretManagerServiceClient()
secret_name = f"projects/{PROJECT_ID}/secrets/{secret_id}/versions/latest"
secret_response = secret_client.access_secret_version(request={"name": secret_name})
pg_password = secret_response.payload.data.decode("UTF-8")
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Define connection parameters
pg_user = "server_role"
pg_host = "35.187.3.132"
pg_port = "5432"
pg_database = "Live"
pg_schema = "l2_stats"
pg_table = "cron"

connection_string = f"postgresql+psycopg2://{pg_user}:{pg_password}@{pg_host}:{pg_port}/{pg_database}"

try:
    engine = create_engine(connection_string)
    df_pg = pd.read_sql_table(pg_table, con=engine, schema=pg_schema)

    display(df_pg)

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    if 'engine' in locals():
        engine.dispose()
```
